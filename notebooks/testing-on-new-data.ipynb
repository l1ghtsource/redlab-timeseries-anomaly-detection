```python
df = pd.read_csv(r'C:\Users\oiica\OneDrive\Рабочий стол\Кейс 1\data.tsv', sep='\t', header=None, names=column_names)

df.head()
```

```python
import pandasql as psql

query = """
select
 point as time,
 sum(total_call_time) * 1.0 / sum(call_count) as "Web Response"
from df
where
 language = 'java'
 and app_name = '[GMonit] Collector'
 and scope = ''
 and name = 'HttpDispatcher'
group by point
order by point
"""

result = psql.sqldf(query, locals())
```
